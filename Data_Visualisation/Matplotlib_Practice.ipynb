## Alttaki görselleri araştırarak yapmaya çalışalım. Zihnimizi ve parmaklarımızı zorlayalım:)

#### Bir hatırlatma "fig, ax = plt.subplots" syntax i size cevabı bulmanızda yardımcı olur. 

#### ayrıca shift + tab+ tab yapmayı unutmayın ve içindeki parametreleri denemeyi unutmayın.

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
# expected output
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 1, figsize= (8.5,6))

ax[0].plot(labels,july16_2007 , color='steelblue',  lw=2)
ax[1].plot(labels, july16_2020, color='steelblue', lw=2)


plt.show()
```

## Bu görselleri alabilenler için alkışı borç bilirim

```python
# expected output1
```

```python
fig, ax1 = plt.subplots(figsize=(8,5))

color = 'blue'
ax1.set_xlabel('Time Period')
ax1.set_ylabel('July 16 2007', color=color)
ax1.plot(labels, july16_2007, color=color)

ax2 = ax1.twinx()  # İkinci bir eksen oluştur
color = 'red'
ax2.set_ylabel('July 16 2020', color=color)
ax2.plot(labels, july16_2020, color=color)

fig.tight_layout()  # Düzeni sıkılaştır
plt.show()
```

```python
#expected output2
```

```python
# Çift eksenli grafik oluşturma
fig, ax1 = plt.subplots(figsize=(9,6))

# İlk eksen için çizgi rengini ayarlama
ax1.spines['left'].set_color('blue')  # Sol eksen çizgisini mavi yap
ax1.spines['right'].set_color('red')  # Sağ eksen çizgisini kırmızı yap

# İkinci eksen için çizgi rengini ayarlama
ax2 = ax1.twinx()
ax2.spines['left'].set_color('blue')  # Sol eksen çizgisini mavi yap
ax2.spines['right'].set_color('red')  # Sağ eksen çizgisini kırmızı yap

# Verileri çizme
ax1.plot(labels, july16_2007, color='blue')
ax2.plot(labels, july16_2020, color='red')

plt.show()
```
