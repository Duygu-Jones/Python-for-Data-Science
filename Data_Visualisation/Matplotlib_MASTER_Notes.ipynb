<font color='hotpink'> <b>Code Aklamas;</b><font color='black'>
    

Bu kod blounda, bir sins dalgas izmek iin Matplotlib kullanlmtr.

1. `x = np.linspace(0, 10, 100)`: Bu komut, 0 ile 10 arasnda eit aralkl 100 say retir. Bu saylar, x-eksenindeki deerlerimizi oluturur.

2. `y = np.sin(x)`: Numpy'nin `sin` fonksiyonunu kullanarak, `x` dizisinin her bir deeri iin sins deerlerini hesaplar. Bu deerler, y-eksenindeki deerlerimizi oluturur.

3. `plt.figure(figsize=(8, 6))`: Grafiin boyutunu 8x6 birim olarak ayarlar. Bu, grafik boyutunu zelletirmek iin kullanlr.

4. `plt.plot(x, y, "r")`: `x` ve `y` dizilerini kullanarak bir izgi grafii izer ve izgi rengini krmz ("r") olarak belirler.

5. `plt.show()`: izilen grafii gsterir. Bu komut, grafiin ekranda grnmesini salar.

````python
x= np.linspace(0,10,100)  # 0 ile 10 arasinda esit uzakliklarda 100 adet nokta-gozlem belirler
y= np.sin(x)              # sin() fonk kullanarak x dizisinin her elemaninin sinusunu hesaplar

plt.figure(figsize=(8,6))  # plt.figure() ile figsize=(tuple) formatinda grafik boyutu belirlenebilir
plt.plot(x,y,'b')

plt.show()
````

##  <font color='Purple'> <b>Object Oriented Method(OOM)</b><font color='black'>   

### <font color='dodgerblue'> <b> fig, ax = plt.subplots(2,1)(Matrix - Multiple Graphs)</b><font color='black'>
 <font color='hotpink'> <b> Code Aklamas: </b><font color='black'>
    
Bu kod blou, Matplotlib ktphanesi kullanarak, iki farkl maa veri setini karlatrmak iin dikey olarak sralanm iki alt grafik ieren bir izgi grafii oluturur. Kod, `subplots` fonksiyonu araclyla daha dzenli ve esnek bir ekilde alt grafik ynetimi salar.

1. `fig, ax = plt.subplots(2, 1)`: Bu komut, iki satr ve bir stun olacak ekilde dzenlenmi iki alt grafii ieren bir figr (`fig`) ve alt grafikler iin eksenler (`ax`) oluturur. `ax` bir dizi olarak dndrlr, her bir eleman bir alt grafiin eksenlerini temsil eder.

2. `ax[0].plot(age, salary, "r")`: Dizinin ilk eleman (`ax[0]`), ya (`age`) ve ilk maa seti (`salary`) verilerini kullanarak krmz (`r`) renkli bir izgi grafii izer. Bu, veri setinin yaa bal maa dalmn temsil eder.

3. `ax[1].plot(age, salary_2, "b")`: Dizinin ikinci eleman (`ax[1]`), ya (`age`) ve ikinci maa seti (`salary_2`) verilerini kullanarak mavi (`b`) renkli bir izgi grafii izer. Bu, farkl bir veri setinin yaa bal maa dalmn gsterir.

4. `plt.tight_layout()`: Bu komut, figrdeki alt grafikler arasndaki boluklar otomatik olarak ayarlar, bylece etiketler, balklar ve eksen isimleri gibi geler st ste binmez ve her ey net bir ekilde grlebilir hale gelir.

5. `plt.show()`: Oluturulan izgi grafiini ekranda gsterir.

````python
fig, ax = plt.subplots(2,1)

ax[0].plot(age, salary, 'r')
ax[1].plot(age, salary_2, 'b')

plt.tight_layout()
plt.show()

````

Bu dzenleme, iki farkl maa veri setinin karlatrlmasn kolaylatrr ve yaa gre deiim trendlerini aka gsterir.

####  <font color='hotpink'> <b> Summary:OOM / Func.Methods </b><font color='black'>
lk kod blou, Matplotlib'in fonksiyonel yaklamn kullanr. Bu yaklamda, plt.subplot fonksiyonu ile dorudan alt grafikler oluturulur ve her bir alt grafik iin plt.plot fonksiyonu arlarak izimler yaplr. Bu yntem, daha basit grselletirmeler iin hzl ve kolay bir yol sunar.
    

    
* kinci kod blou ise Matplotlib'in nesne ynelimli programlama (OOP) yaklamn kullanr. Bu yaklamda, plt.subplots fonksiyonu kullanlarak bir figr ve alt grafiklerin eksenleri (ax) bir kezde oluturulur. Daha sonra, bu eksen objeleri zerinden izimler yaplr. Bu yntem, daha karmak grselletirmeler ve ince ayarlar iin daha fazla kontrol ve esneklik salar.



Her iki yntemin de avantajlar vardr:

* Fonksiyonel yntem, basit grselletirmeler iin hzl ve kolay bir yol sunar.
* OOP yntemi, daha karmak grselletirmelerde ve izimlerde daha iyi bir kontrol ve esneklik salar, ayrca kodun yeniden kullanlabilirliini ve okunabilirliini artrr.

### <font color='dodgerblue'> <b> np.polyfit(x, y, 1): Basit dorusal regresyon hesaplamas </b><font color='orange'>


  <font color='hotpink'> <b> Code Aklamas: </b><font color='black'>

Bu kod blou, Matplotlib ve NumPy ktphaneleri kullanlarak, bir restorann toplam fatura miktarlarna gre braklan bahi miktarlarn gsteren bir salm grafii oluturur ve bu verilere en uygun dorusal regresyon izgisini ekler. Dorusal regresyon izgisi, iki deiken arasndaki ilikiyi daha net anlamamza yardmc olur.

1. `fig, ax = plt.subplots()`: Bir figr ve bir eksen oluturur. Bu yaklam, grafiin eitli zelliklerini daha kolay kontrol etmemizi salar.

2. `ax.scatter(tips.total_bill, tips.tip)`: `scatter` metodu ile `tips` veri setinden `total_bill` (toplam fatura) ve `tip` (bahi) stunlar kullanlarak salm grafii izilir. Her bir nokta, bir restoran mterisinin dedii toplam fatura ve brakt bahi miktarn temsil eder.

3. `x=tips.total_bill` ve `y=tips.tip`: `x` ve `y` deikenleri srasyla `total_bill` ve `tip` stunlarna atanr. Bu, dorusal regresyon hesaplamas iin gerekli olan verileri hazrlar.

4. `m, b = np.polyfit(x, y, 1)`: `np.polyfit` fonksiyonu, `x` ve `y` zerinde basit bir dorusal regresyon (1. dereceden polinom) hesaplar ve modelin eimi (`m`) ve y-intercept'i (`b`) olmak zere iki parametre dndrr. Bu, veri setine en uygun doruyu bulmamz salar.

5. `ax.plot(x, m*x + b, c="r")`: Hesaplanan eim ve y-intercept kullanlarak, salm grafii zerine krmz (`c="r"`) renkte bir dorusal regresyon izgisi izilir. Bu izgi, `tips` veri setindeki `total_bill` ve `tip` arasndaki genel ilikiyi gsterir.

6. `plt.show()`: Oluturulan salm grafii ve zerine eklenen dorusal regresyon izgisini gsterir.

[CueMath Formuller](https://www.cuemath.com/geometry/y-mx-b/)

````python
fig, ax = plt.subplots()

ax.scatter(tips.total_bill, tips.tip)

x=tips.total_bill
y=tips.tip

m, b = np.polyfit(x, y, 1) #Basit dorusal regresyon hesaplamas

ax.plot(x, m*x + b, c="r")

plt.show()
````

  <font color='hotpink'> <b> Output: </b><font color='black'>
    
Bu grafik genel olarak bir veri setinin iki deikeni arasndaki ilikiyi gsteren bir salm (scatter) grafiidir. X eksenindeki deerler arttka Y eksenindeki deerlerin de artma eiliminde olduunu gsteren bir dorusal iliki izlenmektedir. Krmz izgi, veri noktalar arasndaki bu ilikinin en iyi dorusal uyumunu (line of best fit) temsil ediyor, yani bir dorusal regresyon modelinin tahmini sonucu olabilir.

Bu tr bir grafik genellikle korelasyonun grselletirilmesi veya basit dorusal regresyon analizi iin kullanlr. Korelasyon, iki deiken arasndaki ilikinin gcn ve ynn ler. Grafikteki veri noktalarnn dalm ve krmz izginin eimi, deikenler arasnda pozitif bir iliki olduunu gsterir  yani bir deiken arttnda dier deikenin de genellikle artmas beklenir.

Ayrca, veri noktalarnn krmz izgi etrafnda ne kadar sk ya da dank olduuna bakarak, modelin verilere ne kadar iyi uyduu hakknda fikir edinebiliriz. Noktalar izgi etrafnda ne kadar yaknsa, o kadar yksek bir uyum (ve dolaysyla yksek bir korelasyon katsays) olduunu syleyebiliriz. Ancak noktalarn izgi etrafnda olduka geni bir alana yayldn gryoruz, bu da modelin tm varyans aklamadn ve tahminlerin belli bir hata pay iereceini gsterir.

    Bu kod blou, Python'un Matplotlib ktphanesi ve pandas ktphanesi kullanlarak, bir restoranda cinsiyet dalmna gre yemek saysn gsteren bir pasta grafii oluturmak iin yazlmtr. Bu tr bir grselletirme, cinsiyete gre mteri dalmn ve yemek sipari etme eilimlerini analiz etmek iin kullanlabilir.

1. `plt.figure(figsize=(10, 8))`: Pasta grafiinin izilecei figrn boyutunu belirler. Bu, figrn geniliinin 10 in ve yksekliinin 8 in olaca anlamna gelir, bylece grafik detaylar daha net grlebilir.

2. `sex_counts = tips['sex'].value_counts()`: `tips` veri setindeki `sex` stunundaki deerlerin saysn hesaplar. Bu, veri setindeki erkek ve kadn mteri saysn belirler ve sonu, cinsiyet kategorilerine gre sralanm bir pandas Serisi olarak dndrlr.

3. `color_list = ["lightskyblue", "lightcoral"]` ve `explode_list = [0.1, 0]`: Pasta grafiindeki dilimler iin renkleri ve bir dilimin dierlerinden "patlatlarak" vurgulanmasn salayan parametreleri belirler. Bu rnekte, ilk dilim (`Male`) biraz daha vurgulanr.

4. `plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=140, pctdistance=0.5, labeldistance = 0.4, shadow = True, colors = color_list, explode = explode_list)`: zelletirilmi pasta grafiini izer. `sex_counts` verisi dilimlerin byklklerini, `labels` cinsiyet kategorilerini, `autopct` yzdelik deerlerin formatn, `startangle` grafiin balang asn, `pctdistance` ve `labeldistance` yzdelik deerlerin ve etiketlerin dilim merkezine olan uzaklklarn, `shadow` glge efektini, `colors` dilim renklerini ve `explode` belirli dilimlerin vurgulanmasn belirler.

5. `plt.title("Tips Dataset - Meal Counts by Sex")`: Grafie bir balk ekler. Bu balk, grafiin neyi temsil ettiini aka belirtir ve okuyucunun grafii daha kolay anlamasn salar.

````python
plt.figure(figsize=(10, 8))

sex_counts = tips['sex'].value_counts() # 'sex' stununa gre veri saysn hesaplama

color_list = ["lightskyblue", "lightcoral"]
explode_list = [0.1, 0]


plt.pie(sex_counts, labels=sex_counts.index, 
        autopct='%1.1f%%', 
        startangle=140, 
        pctdistance=0.5,
        labeldistance = 0.4,
        shadow = True,
        colors = color_list,
       explode = explode_list)
plt.title("Tips Dataset - Meal Counts by Sex")

plt.show()
````
