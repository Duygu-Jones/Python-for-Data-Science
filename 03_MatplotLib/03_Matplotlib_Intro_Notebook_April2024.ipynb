##  <font color='Purple'> <b>Object Oriented Method(OOM)</b><font color='black'>   

#### <font color='mediumorchid'> <b> fig, ax = plt.subplots(2,1)(Matrix - Multiple Graphs)</b><font color='black'>

```python
fig, ax = plt.subplots(2,1)

ax[0].plot(age, salary, 'r')
ax[1].plot(age, salary_2, 'b')

plt.tight_layout()
plt.show()

```

```python
fig, ax = plt.subplots(1,2)

ax[0].plot(age, salary, 'r')
ax[1].plot(age, salary_2, 'b')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2,3)  # 2 satirlik, 3 sutunluk alt alta yanyana grafikleri bir kerede olusturur oom
 
ax[0][0].plot(age,salary, 'r')
ax[0][1].plot(age,salary_2, 'b')
ax[0][2].plot(age,salary, 'y')

ax[1][0].plot(age,salary, 'k')
ax[1][1].plot(age,salary_2, 'c')
ax[1][2].plot(age,salary, 'g')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(3,2)  # 3 satirlik, 2 sutunluk alt alta yanyana grafikleri bir kerede olusturur oom
 
ax[0][0].plot(age,salary, 'r')
ax[0][1].plot(age,salary_2, 'b')

ax[1][0].plot(age,salary, 'k')
ax[1][1].plot(age,salary_2, 'c')

ax[2][0].plot(age,salary, 'y')
ax[2][1].plot(age,salary, 'g')

plt.tight_layout()
plt.show()
```

##  <font color='blue'> <b>Multiple Plots / OOM- (Toplu tum parametrelerin kullanimi) </b><font color='black'>  

```python
x = np.arange(-50, 50)
y1 = np.power(x, 2)
y2 = np.power(x, 3)
y3 = np.sin(x)
y4 = np.cos(x)
y5 = np.tan(x)
y6 = np.tanh(x)
y7 = np.sinh(x)
y8 = np.cosh(x)
y9 = np.exp(x)

fig1, ax1 = plt.subplots(nrows = 3, ncols = 3, figsize= (20, 20))
ax1[0, 0].plot(x, y1, "tab:blue") # set the color of line chart
ax1[0, 0].set_title("Square Function") # setting title of subplot
ax1[0, 0].set_xlabel("X", fontsize = 18)
ax1[0, 0].set_ylabel("Y", fontsize = 18)

ax1[0, 1].plot(x, y2, "tab:orange") # set the color of line chart
ax1[0, 1].set_title("Cubic Function") # setting title of subplot
ax1[0, 1].set_xlabel("X", fontsize = 18)
ax1[0, 1].set_ylabel("Y", fontsize = 18)

ax1[0, 2].plot(x, y3, "tab:green") # set the color of line chart
ax1[0, 2].set_title("Sine Function") # setting title of subplot
ax1[0, 2].set_xlabel("X", fontsize = 18)
ax1[0, 2].set_ylabel("Y", fontsize = 18)

ax1[1, 0].plot(x, y4, "r-") # set the color of line chart
ax1[1, 0].set_title("Cosine Function") # setting title of subplot
ax1[1, 0].set_xlabel("X", fontsize = 18)
ax1[1, 0].set_ylabel("Y", fontsize = 18)

ax1[1, 1].plot(x, y5, "b-") # set the color of line chart
ax1[1, 1].set_title("Tangent Function") # setting title of subplot
ax1[1, 1].set_xlabel("X", fontsize = 18)
ax1[1, 1].set_ylabel("Y", fontsize = 18)

ax1[1, 2].plot(x, y6, "g-") # set the color of line chart
ax1[1, 2].set_title("Hyperbolic Tangent Function") # setting title of subplot
ax1[1, 2].set_xlabel("X", fontsize = 18)
ax1[1, 2].set_ylabel("Y", fontsize = 18)

ax1[2, 0].plot(x, y7, "m-") # set the color of line chart
ax1[2, 0].set_title("Hyperbolic Sine Function") # setting title of subplot
ax1[2, 0].set_xlabel("X", fontsize = 18)
ax1[2, 0].set_ylabel("Y", fontsize = 18)

ax1[2, 1].plot(x, y8, "y*") # set the color of line chart
ax1[2, 1].set_title("Hyperbolic Cosine Function") # setting title of subplot
ax1[2, 1].set_xlabel("X", fontsize = 18)
ax1[2, 1].set_ylabel("Y", fontsize = 18)

ax1[2, 2].plot(x, y9, "kv") # set the color of line chart
ax1[2, 2].set_title("Exp Function") # setting title of subplot
ax1[2, 2].set_xlabel("X", fontsize = 18)
ax1[2, 2].set_ylabel("Y", fontsize = 18)

plt.show()
```

####  <font color='dodgerblue'> <b>mpl.rcParams.update(mpl.rcParamsDefault)</b><font color='black'> 
    
mpl.rcParams, Matplotlib ktphanesinin izimlerin grnmn zelletirmek iin kullanlan yaplandrma ayarlarn ieren bir nesnedir. "rc" ksaltmas "resource configuration" veya "runtime configuration" anlamna gelir ve Matplotlib'te izimlerin renkleri, izgi kalnlklar, yaz tipleri ve dier zellikleri gibi eitli zelliklerini kontrol etmek iin kullanlr.
    
mpl.rcParams ile bu ayarlar, Matplotlib tarafndan kullanlan varsaylan deerlerin zerine yazlr. Bu nedenle, rcParams zerinde yapacanz herhangi bir deiiklik, bundan sonra oluturulan tm Matplotlib izimlerini etkiler.

### <font color='Magenta'> <b> Pie Chart</b><font color='black'>

```python
x = np.array([25, 25, 25, 25])
mylabels = ["Apples", "Bananas", "Cherries", "Melon"]
```

```python
plt.figure(figsize = (10, 10))

plt.pie(x, labels = mylabels, labeldistance = 0.50, autopct = "%.2f"); #  "%.2f" = float 2 basamakli ondalik kismi
```

```python
z = np.array([16, 21, 45, 19])
mylabels = ["Apples", "Bananas", "Cherries", "Melon"]
```

```python
plt.figure(figsize = (10, 10))
myexplode = [0.1, 0, 0, 0]

plt.pie(z, labels = mylabels, 
        labeldistance = 0.75,
        autopct = "%.1f",
        explode = myexplode,
       startangle= 90);
```

```python
plt.figure(figsize=(10, 8))

color_list= ["gold", "yellowgreen", "lightcoral", "lightskyblue"]
explode_list = [0.5, 0.5, 0.5, 0.5]

y= np.array([25, 35, 25, 15])

mylabels = ["Apples", "Bananas", "Cherries", "Melon"]

plt.pie(y, labels = mylabels, 
        labeldistance = 0.7, 
        autopct = "%.3f", 
        startangle = 90, 
        shadow = True, 
        pctdistance = 0.6,
       colors = color_list,
       explode = explode_list)

plt.legend(loc = (0))
plt.show()
```
