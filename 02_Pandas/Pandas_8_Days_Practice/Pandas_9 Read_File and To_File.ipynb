# Pandas: Read_Excel and To_Excel *Explained*
- read_excel: function in Pandas that is used for importing data from Excel files into a Pandas DataFrame
- to_excel: function in Pandas used to write or export a Pandas DataFrame to an Excel

### Parameter: header

```python
# financial page 3.index satirdan itibaren baslatilmis, headerlari duzgun bir sekilde gormek icin header=satir index no ile tanimlanabilir
df_financials = pd.read_excel(path, sheet_name='Financials', header=3) # header 3.index satiri al 
df_financials
```

### Parameter: nrows

### Parameter: usecols

```python
# ayrica gormek istedigimiz satir adetini / gormek istedigimiz column label'larini belirtebilirz
df_financials = pd.read_excel(path, sheet_name='Financials', header=3, nrows=100, usecols=['Account', 'Business Unit']) # header 3.index satiri al / n adet rows getir/ Account ve BusUnit column larini goster
df_financials
```

### Parameter: index_col

```python
# veya index label'ini istedigimiz bir sutundan secebiliriz
df_financials = pd.read_excel(path, sheet_name='Financials', header=3, index_col='Account') # Account column'u index label olarak set ettik 
df_financials
```

# To_Excel

```python
df_employees.to_excel("1Exported_Excel_data_Employees.xlsx", sheet_name='Employees2', index=False)
#burada olusturdugumuz df den excel dosyasi olusturduk dosya ismini belirledik; --> "1Exported_Excel_data_Employees.xlsx"
# workspace de calsitigimiz klasor icine export eder
# excel icinde sheet1 adindaki page ismini degistirdik; ---> sheet_name='Employees2'
# burada olusturdugumuz df lerde otomatik oalrk index no atamasi yapilir;  bunlari excel'e tekrar tasimamak icin index=False olark atandi
```

```python
# her iki excel file da ayni anda export etmek icin;

with pd.ExcelWriter('1All_Company_Data.xlsx') as writer1:
    df_employees.to_excel(writer1, sheet_name= 'Employees1', index=False)
    df_employees.to_excel(writer1, sheet_name= 'Financials1', index=False)
    
```
