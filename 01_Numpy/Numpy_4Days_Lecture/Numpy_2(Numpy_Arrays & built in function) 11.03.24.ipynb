## <font color='white'> <b>shape & reshape() </b><font color='white'>
*hatirlatma:*

**shape:** 
- shape bir fonk degil attribute tur, array'in kac satir vesutunu oldugu bilgisini dondurur
- Numpy array boyutunu yani kaç satır ve sütundan oluştuğunu gösteren bir tuple döndürür.
- Bu tuple'daki sayılar, dizinin her bir boyutundaki öğe sayısını gösterir. 
- Örneğin, (m, n) tuple'ı m satır ve n sütundan oluşan bir matrisi temsil eder.

**reshape():** 
- Numpy array’i yeniden şekillendirmek istediğimizde yani satır ve sütun sayısını değiştirmek istediğimizde reshape() metodunu kullanabiliriz.
- Diziyi belirtilen yeni şekle getirerek yeni bir ndarray döndürür. Ancak, bu metod orijinal veriyi değiştirmez.
- Belirli boyutlar için "-1" değeri kullanılabilir, bu sayede NumPy otomatik olarak belirtilen diğer boyutlara uygun şekilde bu boyutu hesaplar.

## <font color='blue'> <b>ndim & ndmin  </b><font color='white'>
**ndim**:
- Bir NumPy array'inin (n dimentional array-> kac boyutlu 1D,2D,3D..)boyutunu gösterir.

**ndmin**:
- ndmin, "minimum dimension" ifadesinin kısaltmasıdır ve Türkçede "minimum boyut" anlamına gelir.
- np.array() fonksiyonunda bu parametreyi kullanarak, oluşturulacak dizinin en az kaç boyutlu olması gerektiğini belirtiriz.
- Eğer verilen veri belirtilen boyutun altındaysa, eksik boyutlar otomatik olarak eklenir. 
- Bu, daha yüksek boyutlu bir diziyle işlem yapmak istendiğinde, fakat verinin bu boyutta olmadığında kullanışlıdır.

```python
arr0 = np.array(5)
arr0
```

```python
arr0.ndim
```

```python
arr1 = np.array([1,2])
arr1
```

```python
arr1.ndim   # kac boyutlu oldugunu verir, n-dimensions
```

```python
my_list = list(range(0,24))
```

```python
my_array = np.array(my_list)
my_array
```

```python
my_array.ndim
```

```python
my_array.shape  #tek boyutlularda satir sutun aramaz direk eleman sayisini getirdi
```

##ndmin

tek boyutlu array leri boyutlandirmak icin kullanilir

## <font color='blue'> <b>dtype & astype() </b><font color='white'>

**dtype**:
- bir attribute ve parametre olarak kullanilabilir veya bir verinin dtype bilgisini getirebilir
- dtype (data type), bir ndarray'de saklanan verinin türünü belirtir.
- Bir ndarray oluştururken dtype parametresi kullanılarak veri tipi belirtilebilir. Eğer belirtilmezse, NumPy verilere uygun bir veri tipi seçer.


**astype()**:
- bir fonskiyondur, icine aldigi veri ile islem yapar ve kopyalar
- astype() bir ndarray metodudur ve mevcut bir ndarray'nin veri tipini değiştirmek için kullanılır.
- Bu metod, verinin yeni bir kopyasını oluşturarak yeni veri tipine dönüştürür. Orijinal dizi değişmez.

```python
my_int = np.array([1, 2, 3])
my_int
```

```python
my_int.dtype    # attribute= bilgi verir
```

```python
my_int = np.array([1, 2, 3], dtype= "int64")    #dtype; attribute & parametre
my_int
```

```python
my_float = np.array([1, 2, 3.2, 4])
my_float
```

```python
my_float.dtype
```

```python
my_string = np.array([1, 2, 3, 4.5, "a"])
my_string
```

```python
my_string.dtype
```

```python
my_float2 = np.array([1,2,3], dtype= "float")
my_float2
```

```python
my_float2.astype("int")
```

```python
my_float2
```

```python
new_int = my_float2.astype("int") # astype() fonk olarak kullanilr, dtype'i bool'a cevirir kopya uzerinde
new_int 
```

```python
new_int.dtype   # orjinal data degismez
```

```python
new_bool = my_float2.astype("bool")     # kalici olmasi icin variable olarak tanimlanabilir
new_bool
```

## <font color='blue'> <b>linspace() </b><font color='white'>
Verilen başlangıç ve bitiş değerleri arasında, belirtilen sayıda eşit aralıklı değerlerden oluşan bir numpy dizisi oluşturur.

Bu fonksiyon, özellikle grafik çizme ve veri analizi gibi durumlarda yararlıdır.

Belirlenen [start, stop] degerlerine gore start ve stop'da DAHIL!! edecek sekilde esit uzaklikta noktalara boler

NOTE: aralik sayisi belirtilmez ise default olarak 50 esit parcaya boler(start ve stop dahil !!)



```python
np.linspace(1,2)    # 1 ve 2 sahil 50 esit parca olusturdu- belirtilmez ise vektor(tek boyutlu) array olarak verir
```

```python
# kac parcaya ayiracagini default degerini degistirip deger atayabiliriz
np.linspace(1,2, num=3)     #start+stop dahil  3 parca olusturdu ==> [1.--1.5--2.]
```

```python
#endpoint yani stop pointi dahil etmemek icin default degerini False olarak atayabiliriz
np.linspace(1,2, num=3, endpoint=False)     # stop dahil degil
```

```python
# 0 ve 1 arasinda (0-1 dahil) 30 nokta uretelim;

np.linspace(0,1,30)
```

```python
# 0 ve 1 arasinda (0 dahil, 10 haric) 25 nokta uretelim; steps ler arasi mesafeyi gosterelim;

np.linspace(0,10, num=25, endpoint=False, retstep=True)     # endpoint=False-> stop haric, defaul=True; 
                                                            # reststep=True-> parcalar arasi mesafeyi verir/gosterir, defaul=False'tur,normalde gostermez,
```

```python
np.linspace(1,10,4)
```

```python
np.linspace(1,10,5)
```

```python
np.linspace(1, 20)
```

```python
np.linspace(1, 20, dtype= "int")
```

```python
np.linspace(1,10,4, endpoint=True)
```

```python
np.linspace(1,10,4, endpoint=False)
```

```python
np.linspace(1, 20, 10)
```

```python
np.linspace(1, 10, 4, retstep = False)
```

```python
np.linspace(1, 10, 4, retstep = True)
```

## <font color='blue'> <b>logspace() </b><font color='white'>
logspace() logaritmik aralıklarda bir dizi oluşturur. 

logspace() fonksiyonu, linspace() fonksiyonuna benzer şekilde çalışır, ancak aralıklar logaritmik olarak hesaplanır.

Logspace : Başlangıç değeri 10a ,bitiş değeri değeri 10b olan n elemanlı ve elemanları arasındaki katları eşit olan bir dizi oluşturur.


```python
np.logspace(2,3, num= 5)  # default base = 10     10 ^ 2 = 100, 10 ^3 = 1000
```

```python
177.827941 - 100
```

```python
316.22776602 - 177.827941
```

## <font color='blue'> <b>eye() </b><font color='white'>
- Bu fonksiyon, belirtilen boyutta bir birim matris (identity matrix) oluşturur. 
- Birim matriste, ana köşegen üzerindeki elemanlar 1'dir ve diğer tüm elemanlar 0'dır.

NOTE: 
- eye() = identity= birim matrix demektir
- matrix'lerdeki etkisiz eleman mantiginda calisir
- bir matrix ile eye matrix carptigimizda diogonal degeri 1 olan matrix verir (math islem konusundaki etkisiz eleman gibi)
- esit kenarli bir matrix olmadiginda yine bir kosegeni 1 ile baslar ve capraz olarak 1 ile devam eder, diger tum birimler 0 degerini alir



```python
np.eye(3,4)     # 3'e 4'luk matrix yapini olusturur ve  kosegeni etkisiz eleman 1 olur, diger tum elemanlari 0'dir
```

```python
np.eye(3,3)     # kare matrix te duzgun kosegen cizer
```

```python
np.eye(4)   # tek deger yazildiginda 4'e 4'luk bir matrix olusturur, ve bir ksegenini etkisiz eleman olan 1 ile olusturur
```

```python
np.eye(4, k = 1)    # kosegeni key olarak nereden baslatacagini belirleyebilirz
```

```python
np.eye(4, k = -2)
```

## <font color='white'> <b>rand & randn & randint </b><font color='white'>

(uniform gelme olasiligi: verilen araliktaki her bir sayinin gelme olasiligi esittir)

**np.random.rand():** 0-1 aralığında uniform dağılıma sahip random FLOAT sayılar üretir.

**np.random.randint():** belirtilen aralıkta uniform dagilimda rastgele tamsayılar üretir. Bitiş dahil değil. 

**np.random.randn():** ortalama değeri mean 0 ve standart sapması 1 olan normal dağılıma sahip rastgele sayılar üretir.

**np.random.normal():** mean, standar deviation, arraySize(yani matrix satir ve sutunlari): satir ve sutun sayisina gore mean etrafinda std sapma miktarina gore sayilar uretir --> normal dagilim olusturmak. 

```python
np.random.normal(0,1, (4,5))    # mean'i 0; std 1 olan 4'e 5'lik bir matrix olusturur==> std miktari kadar ort saginda ve olundan uzaklasarak sayi uretir
```

```python
np.random.randint(1,10,(3,4))
```

```python
np.random.randint(1,10, (3,5,5))
```

```python
#[1,10] dahil sayilarin gelme olasiginin esit oldugu uniform dagilim olup olmadigini gosteren algoritma yaziniz

d={}

for _ in range(2000):           # _ altTire ile i'yi kullanmadan loop olusturur
    key = np.random.randint(1,11)
    
    if key not in d:  # eger key d icinde yok ise
        d[key] =1     # d[key] ine=  1 degeri value olarak atanir
    else:
        d[key] +=1    # key, d icinde var ise; d[key] value degeri +1 artirilir
        
d
```
