## <font color='white'> <b>array() </b><font color='white'>

- **NumPy**, bilimsel ve sayısal hesaplamalar için dizilerle -arrays- çalışmayı optimize eden bir Python kütüphanesidir.
- Bu kütüphanede yer alan çok boyutlu dizilere **ndarray** adı verilir. Bu isim, 'n-boyutlu dizi' anlamına gelir.
- **array()** fonksiyonunu kullanarak bir NumPy ndarray nesnesi oluşturabiliriz.

**np.array()** fonksiyonu, Python listelerini, tuple'ları ya da diğer dizi benzeri veri yapılarını NumPy dizisine (ndarray) dönüştürmek için kullanılır. 

**En sık kullanılan parametresi:**

**dtype**:
   - Dizide saklanacak öğelerin veri tipini belirtir.
   - Örnek: float, int, complex. Eğer belirtilmezse, NumPy veri tipini girdi veriye göre otomatik olarak belirler.
   - Array, icersine sadece tek type data alir veya en gecerli data type'a digerlerini caste eder
   - string> float> int seklinde oncelik siralanabilir

```python
my_list = [1, 2, 3]
my_list
```

```python
np.array(my_list)
```

```python
np.array([4, 5, 6.5])   # default(dtype= float) -> type belirtilmez ise numeric datalari float olarak degerlendirir
```

```python
np.array([7, 8, 9], dtype= float)   #dataType = seklinde istelenilen dtype atanarak degistirilebilir
```

```python

#ORN: calculate the miles per gallon for each portion of the trip;
miles_lsit=[280,300,360,400]
gallons_list = [7.1, 8.9, 10.1, 6.8]

# list'lerde toplu olarak bolme islemi yapilamaz bu nedenle numpy array kullanilacak
miles = np.array(miles_lsit)
gallons = np.array(gallons_list)

mpg = miles / gallons
print(mpg)
```

```python
print(mpg[1])
```

```python
print(mpg> 35)      #her degerin bool sonucunu dondurur
```

```python
print(mpg[mpg > 35])    #selection portion of data ->more practical with additional funct provided by numbers.
```

## object dtype
dtype=object, Python'da genellikle NumPy kütüphanesi içinde kullanılır ve bir veri dizisinin (array) veri tipini belirtir. Bu belirteç, dizi içindeki her öğenin bir Python nesnesi olduğunu belirtir.

NumPy, genellikle homojen veri dizileri üzerinde çalışır, yani bir dizi içindeki tüm öğeler aynı veri tipine sahiptir. Ancak, dtype=object kullanıldığında, dizi herhangi bir Python nesnesini (string, liste, sözlük vb.) içerebilir. Bu, NumPy dizisinin her öğenin kendi veri türünü saklamasına izin verir.

Bu durumda, arr adlı NumPy dizisi, her türlü veri türünü içerebilen bir dizi olacaktır. Bu tür diziler genellikle belirli bir veri tipine sahip olmayan veya farklı türde veri tiplerini içeren verilerle çalışırken kullanılır. Ancak, performans açısından homojen veri türlerini kullanan NumPy dizilerine tercih edilmezler.

```python
arr = np.array([1, 'a', [3, 4], {'key': 'value'}], dtype=object)

print(arr)
```

```python
np.array([[1, 2],[3, 4]])
```

```python
my_list2 = [[1, 2, 4], [3, 5, 6], [7, 9, 8], [10, 11, 12]]
my_list2
```

```python
np.array(my_list2)
```

```python
np.array([[[1,2]],[[3,4]]])
```

## <font color='white'> <b>shape & reshape() </b><font color='white'>
**shape:** 
- shape bir fonk degil attribute tur, array'in kac satir vesutunu oldugu bilgisini dondurur
- Numpy array boyutunu yani kaç satır ve sütundan oluştuğunu gösteren bir tuple döndürür.
- Bu tuple'daki sayılar, dizinin her bir boyutundaki öğe sayısını gösterir. 
- Örneğin, (m, n) tuple'ı m satır ve n sütundan oluşan bir matrisi temsil eder.

**reshape():** 
- Numpy array’i yeniden şekillendirmek istediğimizde yani satır ve sütun sayısını değiştirmek istediğimizde reshape() metodunu kullanabiliriz.
- Diziyi belirtilen yeni şekle getirerek yeni bir ndarray döndürür. Ancak, bu metod orijinal veriyi değiştirmez.
- Belirli boyutlar için "-1" değeri kullanılabilir, bu sayede NumPy otomatik olarak belirtilen diğer boyutlara uygun şekilde bu boyutu hesaplar.
​

```python
my_arr = np.array(my_list2)
my_arr
```

```python
my_arr.shape()  #bir fonk degildir sadece satir ve sutun sayisi hakkinda  bilgi dondurur, 
```

```python
my_arr.shape    # 4 satirli , 3 sutunlu bir matrix oldugunu soyler (tuple gorunumunde sonuc dondurur)
```

```python
np.array([1,2,3,4,5]).shape # tek satirli 1D bir yapidir, 1 satir,1sutun var sadece
```

```python
my_arr
```

```python
my_arr.reshape(3,4)      # reshape() bir fonk dur ve verilen datanin eleman sayisina gore istenilen satir ve sutunlari kendisi hesaplr
                        # onemli olan acikla veya fazla eleman kalmamasidir, hesaplamaz hata verir
```

```python
my_arr
```

```python
my_arr.reshape(3,5) # eksik eleman olamaz   --> my-arr list 12 elemanli; reshape ise 3,5 yani 3x5=15 elemanli matrix yapisi olusturabilir, listede bu kadar eleman yok
```

```python
my_arr.reshape(3,3) # fazla eleman kalamaz
```

```python
my_arr
```

```python
my_arr.reshape(6,-1)    # 6 satir olsun; (-1) ise eleman sayisina gore sutunu kendisi belirler 
                        # (eleman sayisini kesin bileedigimiz buyuk datalarda satiri belirleriz sutunu kendisi hesaplayip bize duzenli bir matrix yapisi olsuturur)
```

```python
my_arr.reshape(-1,4)    # tam tersi; -1 satiri temsil eder, eleman sayisina gore kendisi belirler; 4 sutun temsil eder
```

# newaxis attribute
1D boyutlu array'a yeni bir boyut(axis) ekleyebilirz ---> np.newaxis atribute kullanilir

[newaxis atribute'unun bulundugu yeri 1 boyut : diger tarafi ise eleman sayisina(n) gore cogaltir]

data[np.newaxis, :] -> 1 satirli : n sutunlu

data[:, np.newaxis]  -> n satirli : 1 sutunlu matrix haline getirir

```python
a = np.arange(1,11)     # 1D tek boyutlu bir array
a
```

```python
a[np.newaxis, :] # [1 row : 10 column] seklinde eleman sayisina gore boyut ekler
```

```python
a[np.newaxis, :].shape
```

```python
a[:, np.newaxis] # [10 row : 1 column] seklinde eleman sayisina gore boyut ekler
```

```python
a[:, np.newaxis].shape
```
